```python
from collections import OrderedDict
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def DOS(filename, LDOS = False, atom_list=[], offset=0):
    # only support spin polarized calculation
    # LDOS atom number start from 1
    # dos_sum, dos_s, dos_p, dos_d are dos from atom list
    with open(filename, mode = 'r') as fp:
        dos = fp.readlines()
    num_of_state = int(dos[5].split()[2])
    fermi_energy = float(dos[5].split()[3]) 
    total_dos_block = dos[6 : 6+num_of_state]
    total_dos_spin_up = np.array([ map(float, i.split())  for i in total_dos_block])[:, 1]
    total_dos_spin_down = np.array([ map(float, i.split())  for i in total_dos_block])[:, 2]
    dos_energy = np.array([ map(float, i.split())  for i in total_dos_block])[:, 0] - fermi_energy + offset
    total_dos = total_dos_spin_up + total_dos_spin_down
    if not LDOS:
        return dos_energy, total_dos
        #         return dos_energy, total_dos_spin_up, total_dos_spin_down, total_dos
    else:
        dos_sum = np.zeros(num_of_state)
        dos_s = np.zeros(num_of_state)
        dos_p = np.zeros(num_of_state)
        dos_d = np.zeros(num_of_state)
        for atom_num in atom_list:
            dos_atom_num = dos[6+atom_num*(num_of_state+1) : 5+(atom_num+1)*(num_of_state+1)]
            dos_atom_num = np.array([ map(float, i.split())  for i in dos_atom_num])
            dos_sum = dos_sum +  np.sum(dos_atom_num[:, 1:], axis = 1)
            dos_s = dos_s + np.sum(dos_atom_num[:, 1:3], axis = 1)
            dos_p = dos_p + np.sum(dos_atom_num[:, 3:9], axis = 1)
            dos_d = dos_d + np.sum(dos_atom_num[:, 9:19], axis = 1)
        return dos_energy, total_dos, dos_sum, dos_s, dos_p, dos_d
    
    
def get_relax_atom_list(filename):
    poscar = open(filename, 'r')
    pos_body = poscar.readlines()
    poscar.close()
    atom = pos_body[6]
    atom = atom.strip().split()
    num_Ti = int(atom[0])
    num_O = int(atom[1])
    num_total = num_O + num_Ti
    pos_body = pos_body[9: 9+num_total]
    pos_body = [i.strip().split() for i in pos_body] 
    relax_O_list = []
    relax_Ti_list = []
    fixed_list = []
    for i in range(num_total):
        if pos_body[i][-1] == 'F':
            fixed_list.append(i+1)
        elif i+1 <= num_Ti:
            relax_Ti_list.append(i+1)
        else:
            relax_O_list.append(i+1)
    return num_Ti, num_O, num_total, relax_Ti_list, relax_O_list, fixed_list


def get_atom_index_in_layer(POSCAR_file, layer_interest):
    #  index start from 1, the bottom layer is the 1th layer
    layer_interest = np.array(layer_interest)
    layer_interest -= 1
    with open(POSCAR_file) as fp:
        lines = fp.readlines()
    natoms = np.sum(map(int, lines[6].split()))
    if lines[7].strip()[0] == 's' or lines[7].strip()[0] == 'S':
        xyzlines = lines[9: 9+natoms]
    else:
        xyzlines = lines[8: 8+natoms]
    z_list = [float(i.split()[2])  for i in xyzlines]
    _, atom_idx = np.unique(z_list, return_inverse=True)
    mask = np.array([idx in layer_interest for idx in atom_idx])
    return np.arange(len(z_list))[mask] + 1
```
